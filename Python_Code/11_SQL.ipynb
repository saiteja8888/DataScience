```python
import mysql.connector as conn
mydb = conn.connect(host='localhost',user='root',passwd='root')
cursor = mydb.cursor()
```

```python
#2 . create a database called as cardataset
cursor.execute("CREATE DATABASE CarDataSet")
```

```python
cursor.execute("SHOW DATABASES")
cursor.fetchall()
```

```python
#3 . create a table called as car with a column name given in dataset description
cursor.execute("""
CREATE TABLE IF NOT EXISTS cardataset.car(
buying varchar(30),
maint varchar(30),
doors varchar(30),
persons varchar(30),
lug_boot varchar(30),
safety varchar(30),
classvalues varchar(30)
)
""")
```

```python
cursor.execute("USE cardataset")
```

```python
#4 . Dump all the data into car table 
import csv
with open("11_SQL/car.data", "r") as f:
    car_data = csv.reader(f, delimiter = '\n')
    for i in car_data:
        print(i)
        cursor.execute(f"insert into car values({str(i[0])})")
mydb.commit()
```
